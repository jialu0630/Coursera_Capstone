```python
import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
tor_df = pd.DataFrame(data[0])
tor_df.head()
```

Ignore cells with a borough that is <b>Not assigned</b>

```python
tor_df = tor_df[tor_df.Borough != 'Not assigned']
tor_df = tor_df.sort_values(by=['Postal Code','Borough'])

tor_df.reset_index(inplace = True)
tor_df.drop('index',axis = 1, inplace = True)

tor_df.head(20)
```
